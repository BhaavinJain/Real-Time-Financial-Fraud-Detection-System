```python
numerical_col_missing = df[missing_mean.index].select_dtypes(include = np.number).columns.tolist()
categorical_col_missing = df[missing_mean.index].select_dtypes(include = 'object').columns.tolist()

print(f"Number of numerical columns: {len(numerical_col_missing)}")
print(f"Number of categorical columns: {len(categorical_col_missing)}")
```

Forming Imputation Strategy 

1. Categorical Columns : Fill with 'Unknown'
2. D-Columns (Time deltas) : Fill with -999 (They represent days between previous transactions , Missing values likely indicate unavailable historical information)
3. M- Columns (Match Flags , Match : name on cards/address) : Fill with Missing
4. V-Columns : Vesta Engineered features , already engineered features , hence cant do median imputation , would have to replace by -999

```python
categorical_col_missing
```

```python
numerical_col_missing
```

```python
# 1. Filling M columns missing values with 'missing'
m_cols = []
for c in categorical_col_missing:
    if c.startswith('M'):
        m_cols.append(c)

for col in m_cols:
    df[col] = df[col].fillna('missing')

# 2. Remaining categorical columns missing values filled with unkown
not_m_cols = []
for c in categorical_col_missing:
    if not c.startswith('M'):
        not_m_cols.append(c)

for col in not_m_cols:
    df[col] = df[col].fillna('unknown')

# 3. All numeric columns being filled with -999
for col in numerical_col_missing:
    df[col] = df[col].fillna(-999)
```

Creating hour column

```python
# Suppose a transaction occurs at TransactionDT = 100000 (this much time after a reference point)
# So convert to hours = 100000 / 3600 = 27.78 and then take modulo of 24 = 27.78 % 24 = 3.78
# So transaction occurred around 3:47 AM relative to the reference date

df['hour'] = (df['TransactionDT'] / 3600) % 24  

# Cyclic Encoding for Linear models and Autoencoder
df['hour_sin'] = np.sin(2 * np.pi * df['hour'] / 24)
df['hour_cos'] = np.cos(2 * np.pi * df['hour'] / 24)

df.drop(columns = 'TransactionDT' , inplace = True)
df
```
